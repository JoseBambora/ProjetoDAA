```python
import matplotlib.pyplot as plt
def plot_grafico(label,lista_metricas,lista_valores,ax,color):
    lista_tuplos = zip(lista_metricas,lista_valores)
    ax.bar(lista_metricas, lista_valores, color=color, label=label,align='center')
    for metric, value in lista_tuplos:
        ax.text(metric , value, f'{value:.2f}', ha='center', va='bottom', color='black')
    

def grafico(accuracies):
    fig, ax = plt.subplots(figsize=(12, 9))
    list1 = list(accuracies.keys())
    list2 = list(accuracies.values())
    plot_grafico('Models',list1,list2,ax,'red')
    plt.tight_layout()
    plt.show()
```

```python
grafico(accuracies)
```
